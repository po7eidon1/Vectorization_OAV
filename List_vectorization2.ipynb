```python
text_Artem = """Оптимальная мотивация
У вас устойчивый и управляемый интерес к учёбе. Иначе говоря, ваши «хочу» и «надо» совпадают. Это отлично! Лучшие результаты получаются, когда мы учимся с интересом, осознаём ценность знаний и можем применять их на практике. Это как раз про вас. Так держать!

Ваша способность учиться на базовом уровне
Скорее всего, вы хорошо организуете учебный процесс, можете вписать занятия в график и обеспечить себе подходящую среду.
Это отлично, но в стрессовых ситуациях эти навыки могут выключаться. Например, когда вы ошибаетесь по невнимательности и получаете плохую для себя оценку. Вследствие таких неудач даже может появиться желание бросить учёбу.

Вы стремитесь к лучшим результатам, но вам трудно продолжать, когда появляется риск неудачи
Похоже, для вас нередка ситуация, когда вы можете отказаться от каких-то своих целей, если они кажутся вам слишком сложными. Часто это происходит так:
Вы недовольны уровнем своих достижений или знаний в новой теме.
Поэтому когда вы ставите цели, то опираетесь на искаженную идеальную картину себя, который умеет больше, чем на самом деле.
Этот паттерн ведёт к тому, что вам часто приходится прикладывать много сил, а радости от результата вы не испытываете, так как переоцениваете свои возможности.

Вы ожидаете, что преподаватель будет мотивировать вас учиться
Главное для вас в преподавателе — чтобы он вдохновлял и вовлекал в учебный процесс. Это хороший подход, мотивация крайне важна в обучении. Но исходить она должна не только извне, но и изнутри. Чтобы сесть за то самое сложное задание, придётся договариваться с собой."""
```

```python
import spacy
nlp = spacy.load("ru_core_news_md")
```

```python
doc_Artem = nlp(text_Artem)
vec_Artem = doc_Artem.vector
```

```python
stop_words = nlp.Defaults.stop_words
stop_words_list = sorted(list(stop_words))
```

```python
filtered_words = [token.text for token in doc_Artem if not token.is_stop]
filtered_text = ' '.join(filtered_words)
```

```python
print(filtered_text)
```

```python
doc_filtered = nlp(filtered_text)
```

```python
sentences = list(doc_filtered.sents)
for sentence in sentences:
    doc_sentences = nlp(sentence.text)
    #print(sentence.text)
    print("Результат стравнения изначального текста с отфильтрованным ="
          , doc_Artem.similarity(doc_sentences))
```
